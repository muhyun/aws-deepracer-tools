```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')

fig = plt.figure()
new_figsize = fig.get_size_inches() * 1.5
```

```python
def connectpoints(i,p1,p2):
    (x1, y1) = p1
    (x2, y2) = p2
    plt.plot(x1,y1,'ro')
    plt.text(x1,y1,i,fontsize=12)
    plt.plot([x1,x2],[y1,y2],'b--')
```

```python
def visualize_waypoints(waypoints):

    plt.figure(figsize=(new_figsize[0], new_figsize[1]))
    for wp in np.arange(0,len(waypoints)-1):
        connectpoints(wp, waypoints[wp], waypoints[wp+1])
```
